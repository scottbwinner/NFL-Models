Scraping QB Depth Chart Data

```python
import time
from bs4 import BeautifulSoup
import requests

# URL of the page
base_url = 'https://www.ourlads.com/nfldepthcharts/archive/'
numbers = {'288': '2024-25','275':'2023-24','262':'2022-23', '249':'2021-22', '236':'2020-21', '224':'2019-20', 
           '212':'2018-19', '199':'2017-18', '187':'2016-17', '175':'2015-16', '162':'2014-15'}
teams = ['ARZ', 'ATL', 'BAL', 'BUF', 'CAR', 'CHI', 'CIN', 'CLE', 'DAL', 'DEN', 'DET', 'GB', 'HOU', 'IND', 'JAX', 'KC',
         'LAC', 'LAR', 'LV', 'MIA', 'MIN', 'NE', 'NO', 'NYG', 'NYJ', 'PHI', 'PIT', 'SEA', 'SF', 'TB', 'TEN', 'WAS']

qb_data = {}

# Send a GET request to the URL
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

for number in numbers:
    for team in teams:
        # Build URL for year/team
        url = base_url + number + '/' + team

        # Wait between requests
        time.sleep(2)
        response = requests.get(url, headers=headers)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Parse the HTML content
            soup = BeautifulSoup(response.content, 'html.parser')

            # Find the <td> element with class "row-dc-grey" containing "QB"
            qb_td = soup.find('td', class_='row-dc-grey', string='QB')
            #print(qb_td)
            if qb_td != None:
                # Navigate to the next <td> element to get the QB's name
                qb_name_td = qb_td.find_next_sibling('td')
                qb_name_td = qb_name_td.find_next_sibling('td')
                #print(qb_name_td)
                if qb_name_td:
                    qb_name = qb_name_td.get_text(strip=True)
                    print("QB:", qb_name)
                    key = team + ',' + numbers[number]
                    qb_data[key] = qb_name

                else:
                    print("QB name not found")
            else:
                qb_td = soup.find('td', class_='row-dc-wht', string='QB')
                #print(qb_td)
                if qb_td != None:
                    # Navigate to the next <td> element to get the QB's name
                    qb_name_td = qb_td.find_next_sibling('td')
                    qb_name_td = qb_name_td.find_next_sibling('td')
                    #print(qb_name_td)
                    if qb_name_td:
                        qb_name = qb_name_td.get_text(strip=True)
                        print("QB:", qb_name)
                        key = team + ',' + numbers[number]
                        qb_data[key] = qb_name
                else:
                    print("QB field not found")
        else:
            print("Failed to retrieve the webpage")
```

Print out the name, team, and year from each pulled starting qb

```python
for qb in qb_data:
    print(qb + " " + qb_data[qb])
```

Extract the clean name from the scraped data

```python
for qb in qb_data:
    names = qb_data[qb].split(",")
    last_name = names[0]
    first_name = names[1].strip()
    first_name = first_name.split(" ")[0]
    qb_data[qb] = first_name + " " + last_name
```

Change all of the names in the data to be lowercase so that we have universal lowercase for the join

```python
for qb in qb_data:
    qb_data[qb] = qb_data[qb].lower()
    print(qb + " " + qb_data[qb])
```

Change names that are different in our other databases

```python
name_key = {'patrick mahomes ii':'patrick mahomes', 'aj mccarron':'a.j. mccarron', 
            'robert griffin':'robert griffin iii', 'ej manuel':'e.j. manuel', 
            'gardner minshew ii':'gardner minshew'}

for key in name_key:
    for qb in qb_data:
        qb_data[qb] = qb_data[qb].replace(key, name_key[key])
```

Write data to qb_depth_chart.csv

```python
import csv

field_names = ['Team', 'Season', 'Name']
data = []
for qb in qb_data:
    data.append([qb.split(",")[0], qb.split(",")[1], qb_data[qb]])

print(data)

with open('processed_qb_data/qb_depth_chart.csv', 'w') as f:
     
    # using csv.writer method from CSV package
    write = csv.writer(f)
    write.writerow(field_names)
    write.writerows(data)
```
