```python
pfr_tm_def_14_15 = pd.read_csv('../fantasy_data/pfr_tm_def/pfr_tm_def_14-15.csv')
pfr_tm_def_15_16 = pd.read_csv('../fantasy_data/pfr_tm_def/pfr_tm_def_15-16.csv')
pfr_tm_def_16_17 = pd.read_csv('../fantasy_data/pfr_tm_def/pfr_tm_def_16-17.csv')
pfr_tm_def_17_18 = pd.read_csv('../fantasy_data/pfr_tm_def/pfr_tm_def_17-18.csv')
pfr_tm_def_18_19 = pd.read_csv('../fantasy_data/pfr_tm_def/pfr_tm_def_18-19.csv')
pfr_tm_def_19_20 = pd.read_csv('../fantasy_data/pfr_tm_def/pfr_tm_def_19-20.csv')
pfr_tm_def_20_21 = pd.read_csv('../fantasy_data/pfr_tm_def/pfr_tm_def_20-21.csv')
pfr_tm_def_21_22 = pd.read_csv('../fantasy_data/pfr_tm_def/pfr_tm_def_21-22.csv')
pfr_tm_def_22_23 = pd.read_csv('../fantasy_data/pfr_tm_def/pfr_tm_def_22-23.csv')
pfr_tm_def_23_24 = pd.read_csv('../fantasy_data/pfr_tm_def/pfr_tm_def_23-24.csv')
pfr_tm_def_24_25 = pd.read_csv('../fantasy_data/pfr_tm_def/pfr_tm_def_24-25.csv')

pfr_tm_def_14_15['Season'] = '2014-15'
pfr_tm_def_15_16['Season'] = '2015-16'
pfr_tm_def_16_17['Season'] = '2016-17'
pfr_tm_def_17_18['Season'] = '2017-18'
pfr_tm_def_18_19['Season'] = '2018-19'
pfr_tm_def_19_20['Season'] = '2019-20'
pfr_tm_def_20_21['Season'] = '2020-21'
pfr_tm_def_21_22['Season'] = '2021-22'
pfr_tm_def_22_23['Season'] = '2022-23'
pfr_tm_def_23_24['Season'] = '2023-24'
pfr_tm_def_24_25['Season'] = '2024-25'

pfr_tm_def = pd.concat([pfr_tm_def_14_15, pfr_tm_def_15_16, pfr_tm_def_16_17, pfr_tm_def_17_18, pfr_tm_def_18_19, pfr_tm_def_19_20, pfr_tm_def_20_21, pfr_tm_def_21_22, pfr_tm_def_22_23, pfr_tm_def_23_24, pfr_tm_def_24_25])

# Getting rid of statistical summary rows that exist amongst the individual team stats
pfr_tm_def = pfr_tm_def[~pfr_tm_def['Tm'].isin(['Avg Team', 'League Total', 'Avg Tm/G'])]

pfr_schedule_14_15 = pd.read_csv('../fantasy_data/pfr_schedule/pfr_schedule_14-15.csv')
pfr_schedule_15_16 = pd.read_csv('../fantasy_data/pfr_schedule/pfr_schedule_15-16.csv')
pfr_schedule_16_17 = pd.read_csv('../fantasy_data/pfr_schedule/pfr_schedule_16-17.csv')
pfr_schedule_17_18 = pd.read_csv('../fantasy_data/pfr_schedule/pfr_schedule_17-18.csv')
pfr_schedule_18_19 = pd.read_csv('../fantasy_data/pfr_schedule/pfr_schedule_18-19.csv')
pfr_schedule_19_20 = pd.read_csv('../fantasy_data/pfr_schedule/pfr_schedule_19-20.csv')
pfr_schedule_20_21 = pd.read_csv('../fantasy_data/pfr_schedule/pfr_schedule_20-21.csv')
pfr_schedule_21_22 = pd.read_csv('../fantasy_data/pfr_schedule/pfr_schedule_21-22.csv')
pfr_schedule_22_23 = pd.read_csv('../fantasy_data/pfr_schedule/pfr_schedule_22-23.csv')
pfr_schedule_23_24 = pd.read_csv('../fantasy_data/pfr_schedule/pfr_schedule_23-24.csv')
pfr_schedule_24_25 = pd.read_csv('../fantasy_data/pfr_schedule/pfr_schedule_24-25.csv')

pfr_schedule_14_15['Season'] = '2014-15'
pfr_schedule_15_16['Season'] = '2015-16'
pfr_schedule_16_17['Season'] = '2016-17'
pfr_schedule_17_18['Season'] = '2017-18'
pfr_schedule_18_19['Season'] = '2018-19'
pfr_schedule_19_20['Season'] = '2019-20'
pfr_schedule_20_21['Season'] = '2020-21'
pfr_schedule_21_22['Season'] = '2021-22'
pfr_schedule_22_23['Season'] = '2022-23'
pfr_schedule_23_24['Season'] = '2023-24'
pfr_schedule_24_25['Season'] = '2024-25'

pfr_schedule = pd.concat([pfr_schedule_14_15, pfr_schedule_15_16, pfr_schedule_16_17, pfr_schedule_17_18, pfr_schedule_18_19, pfr_schedule_19_20, pfr_schedule_20_21, pfr_schedule_21_22, pfr_schedule_22_23, pfr_schedule_23_24, pfr_schedule_24_25])

pfr_schedule = pfr_schedule[~pfr_schedule['Week'].isin(['WildCard', 'Division', 'ConfChamp', 'SuperBowl'])]

# PFR Full Name -> Standard
pfr_team_name_key = {'Arizona Cardinals':'ARI', 'Atlanta Falcons':'ATL', 'Baltimore Ravens': 'BAL', 'Buffalo Bills': 'BUF', 'Carolina Panthers':'CAR',
                      'Chicago Bears':'CHI', 'Cincinnati Bengals':'CIN', 'Cleveland Browns':'CLE', 'Dallas Cowboys':'DAL', 'Denver Broncos':'DEN', 
                      'Detroit Lions': 'DET', 'Green Bay Packers':'GB', 'Houston Texans':'HOU', 'Indianapolis Colts':'IND', 'Jacksonville Jaguars':'JAX', 
                      'Kansas City Chiefs':'KC', 'Miami Dolphins':'MIA', 'Minnesota Vikings':'MIN', 'New England Patriots':'NE','New Orleans Saints':'NO', 
                      'New York Giants':'NYG', 'New York Jets':'NYJ', 'Oakland Raiders':'LV', 'Las Vegas Raiders':'LV', 'Philadelphia Eagles':'PHI', 
                      'Pittsburgh Steelers':'PIT', 'San Diego Chargers':'LAC', 'Los Angeles Chargers':'LAC', 'San Francisco 49ers':'SF', 'Seattle Seahawks':'SEA', 
                      'St. Louis Rams':'LAR', 'Los Angeles Rams':'LAR', 'Tampa Bay Buccaneers':'TB', 'Tennessee Titans':'TEN', 'Washington Redskins':'WAS', 
                      'Washington Commanders':'WAS', 'Washington Football Team':'WAS'}


for key in pfr_team_name_key:
    pfr_tm_def = pfr_tm_def.replace(key, pfr_team_name_key[key])
    pfr_schedule = pfr_schedule.replace(key, pfr_team_name_key[key])


teams = pfr_tm_def['Tm'].unique()
seasons = pfr_tm_def['Season'].unique()

# Creating rankings for best team run defenses

pfr_tm_def['rec_def_points'] = pfr_tm_def['Pass_Cmp'] + (pfr_tm_def['Pass_Yds'] / 10) + (pfr_tm_def['Pass_TD'] * 6)

# Points Allowed Ranking
for season in seasons:
    pfr_tm_def_season = pfr_tm_def[pfr_tm_def['Season'] == season]
    pfr_tm_def_season['points_allowed_rank'] = pfr_tm_def_season['PA'].rank(ascending=True, method='min')
    pfr_tm_def.loc[pfr_tm_def['Season'] == season, 'points_allowed_rank'] = pfr_tm_def_season['points_allowed_rank'].values
    pfr_tm_def_season['rec_def_rank'] = pfr_tm_def_season['rec_def_points'].rank(ascending=True, method='min')
    pfr_tm_def.loc[pfr_tm_def['Season'] == season, 'rec_def_rank'] = pfr_tm_def_season['rec_def_rank'].values


# Aggregating team schedules for each season and calculating average opponent defense rankings
pfr_sos = []
for season in seasons:
    for team in teams:
        wins = pfr_schedule.loc[(pfr_schedule['Winner/tie'] == team) & (pfr_schedule['Season'] == season), ['Winner/tie', 'Loser/tie', 'Season']]
        losses = pfr_schedule.loc[(pfr_schedule['Loser/tie'] == team) & (pfr_schedule['Season'] == season), ['Winner/tie', 'Loser/tie', 'Season']]
        opponents = []
        opponents.extend(wins['Loser/tie'].tolist())
        opponents.extend(losses['Winner/tie'].tolist())
        average_opponent_def_rank = []
        average_opponent_rec_def_rank = []
        for opponent in opponents:
            average_opponent_def_rank.append(pfr_tm_def[(pfr_tm_def['Tm'] == opponent) & (pfr_tm_def['Season'] == season)]['points_allowed_rank'].item())
            average_opponent_rec_def_rank.append(pfr_tm_def[(pfr_tm_def['Tm'] == opponent) & (pfr_tm_def['Season'] == season)]['rec_def_rank'].item())
        
        pfr_sos.append({
            'Tm': team,
            'Season': season,
            'average_opponent_def_rank': sum(average_opponent_def_rank) / len(average_opponent_def_rank),
            'average_opponent_rec_def_rank': sum(average_opponent_rec_def_rank) / len(average_opponent_rec_def_rank)
        })
        
pfr_sos = pd.DataFrame(pfr_sos)

pfr_sos
```
