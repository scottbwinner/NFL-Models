Scraping rb Depth Chart Data

```python
import time
from bs4 import BeautifulSoup
import requests

# URL of the page
base_url = 'https://www.ourlads.com/nfldepthcharts/archive/'
numbers = {'288': '2024-25','275':'2023-24','262':'2022-23', '249':'2021-22', '236':'2020-21', '224':'2019-20', 
           '212':'2018-19', '199':'2017-18', '187':'2016-17', '175':'2015-16', '162':'2014-15'}
teams = ['ARZ', 'ATL', 'BAL', 'BUF', 'CAR', 'CHI', 'CIN', 'CLE', 'DAL', 'DEN', 'DET', 'GB', 'HOU', 'IND', 'JAX', 'KC',
         'LAC', 'LAR', 'LV', 'MIA', 'MIN', 'NE', 'NO', 'NYG', 'NYJ', 'PHI', 'PIT', 'SEA', 'SF', 'TB', 'TEN', 'WAS']

rb_data = {}

# Send a GET request to the URL
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

for number in numbers:
    for team in teams:
        # Build URL for year/team
        url = base_url + number + '/' + team

        # Wait between requests
        time.sleep(2)
        response = requests.get(url, headers=headers, timeout=(15,60))

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Parse the HTML content
            soup = BeautifulSoup(response.content, 'html.parser')

            # Find the <td> element with class "row-dc-wht" containing "RB"
            rb_td = soup.find('td', class_='row-dc-grey', string='RB')
            #print(rb_td)
            if rb_td != None:
                #print(rb_name_td)
                i = 1
                while i < 5:
                    # Navigate to the next <td> element to get the RB's name
                    rb_td = rb_td.find_next_sibling('td')
                    rb_td = rb_td.find_next_sibling('td')
                    if rb_td:
                        rb_name = rb_td.get_text(strip=True)
                        if len(rb_name) > 0:
                            print("RB", i, ": ", rb_name)
                            key = team + ',' + numbers[number] + ',RB' + str(i)
                            rb_data[key] = rb_name
                    else:
                        print("rb name not found")
                    i += 1

            else:
                rb_td = soup.find('td', class_='row-dc-wht', string='RB')
                #print(rb_td)
                if rb_td != None:
                    #print(rb_name_td)
                    i = 1
                    while i < 5:
                        # Navigate to the next <td> element to get the RB's name
                        rb_td = rb_td.find_next_sibling('td')
                        rb_td = rb_td.find_next_sibling('td')
                        if rb_td:
                            rb_name = rb_td.get_text(strip=True)
                            if len(rb_name) > 0:
                                print("RB", i, ": ", rb_name)
                                key = team + ',' + numbers[number] + ',RB' + str(i)
                                rb_data[key] = rb_name
                        else:
                            print("rb name not found")
                        i += 1
                else:
                    print("rb field not found")
        else:
            print("Failed to retrieve the webpage")
```

Print out the name, team, and year from each pulled starting rb

```python
for rb in rb_data:
    print(rb + " " + rb_data[rb])
```

Extract the clean name from the scraped data

```python
for rb in rb_data:
    names = rb_data[rb].split(",")
    last_name = names[0]
    first_name = names[1].strip()
    first_name = first_name.split(" ")[0]
    rb_data[rb] = first_name + " " + last_name
```

Change all of the names in the data to be lowercase so that we have universal lowercase for the join

```python
for rb in rb_data:
    rb_data[rb] = rb_data[rb].lower()
    print(rb + " " + rb_data[rb])
```

Change names that are different in our other databases

```python
name_key = {'tyrone tracy jr.':'tyrone tracy', 'patrick taylor jr.':'patrick taylor', 
            'brian robinson jr.':'brian robinson', 'darrell henderson jr.':'darrell henderson', 
            'demetric felton jr.':'demetric felton', 'melvin gordon': 'melvin gordon iii', 
            'cj anderson': 'c.j. anderson', 'mark ingram': 'mark ingram ii', 'bobby rainey': 'bobby rainey jr.',
            'cj spiller': 'c.j. spiller', 'todd gurley': 'todd gurley ii', 'duke johnson': 'duke johnson jr.', 
            'cj prosise': 'c.j. prosise', 'tj yeldon': 't.j. yeldon', "jd mckissic": "j.d. mckissic", 
            "dj foster": "d.j. foster", "john kelly": "john kelly jr.", "de'angelo henderson": "de'angelo henderson sr.", 
            "tj logan": "t.j. logan", "ronald jones ii": "ronald jones", "michael boone": "mike boone", 
            "jeffery wilson": "jeff wilson jr.", "lamical perine": "la'mical perine", "jk dobbins": "j.k. dobbins", 
            "demetric felton": "demetric felton jr."}

for key in name_key:
    for rb in rb_data:
        rb_data[rb] = rb_data[rb].replace(key, name_key[key])
```

Write data to rb_depth_chart.csv

```python
import csv

field_names = ['Team', 'Season', 'Depth Chart', 'Name']
data = []
for rb in rb_data:
    data.append([rb.split(",")[0], rb.split(",")[1], rb.split(",")[2], rb_data[rb]])

print(data)

with open('processed_rb_data/rb_depth_chart.csv', 'w') as f:
     
    # using csv.writer method from CSV package
    write = csv.writer(f)
    write.writerow(field_names)
    write.writerows(data)
```
