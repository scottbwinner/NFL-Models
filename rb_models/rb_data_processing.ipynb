Reading in pff_rushing data

```python
import pandas as pd

# Reading in pff_rushing data

pr14_15 = pd.read_csv('../fantasy_data/pff_rushing/pff_rushing_14-15.csv')
pr15_16 = pd.read_csv('../fantasy_data/pff_rushing/pff_rushing_15-16.csv')
pr16_17 = pd.read_csv('../fantasy_data/pff_rushing/pff_rushing_16-17.csv')
pr17_18 = pd.read_csv('../fantasy_data/pff_rushing/pff_rushing_17-18.csv')
pr18_19 = pd.read_csv('../fantasy_data/pff_rushing/pff_rushing_18-19.csv')
pr19_20 = pd.read_csv('../fantasy_data/pff_rushing/pff_rushing_19-20.csv')
pr20_21 = pd.read_csv('../fantasy_data/pff_rushing/pff_rushing_20-21.csv')
pr21_22 = pd.read_csv('../fantasy_data/pff_rushing/pff_rushing_21-22.csv')
pr22_23 = pd.read_csv('../fantasy_data/pff_rushing/pff_rushing_22-23.csv')
pr23_24 = pd.read_csv('../fantasy_data/pff_rushing/pff_rushing_23-24.csv')
pr24_25 = pd.read_csv('../fantasy_data/pff_rushing/pff_rushing_24-25.csv')

# Creating a season column to differentiate between seasons

pr14_15['Season'] = '2014-15'
pr15_16['Season'] = '2015-16'
pr16_17['Season'] = '2016-17'
pr17_18['Season'] = '2017-18'
pr18_19['Season'] = '2018-19'
pr19_20['Season'] = '2019-20'
pr20_21['Season'] = '2020-21'
pr21_22['Season'] = '2021-22'
pr22_23['Season'] = '2022-23'
pr23_24['Season'] = '2023-24'
pr24_25['Season'] = '2024-25'

# Combining all of the different season's data into one pff_rushing dataset

pff_rushing = pd.concat([pr14_15, pr15_16, pr16_17, pr17_18, pr18_19, pr19_20, pr20_21, pr21_22, pr22_23, pr23_24, pr24_25])

# Filter on only Running Backs

pff_rushing = pff_rushing[(pff_rushing['position'] == 'HB') | (pff_rushing['position'] == 'FB')]

# Make a column for lowercase player name to join with our depth chart scraped data

pff_rushing['lower_name'] = pff_rushing['player'].apply(lambda x: x.lower() if isinstance(x, str) else x)

# Read in rb_depth_chart data

rb_depth_chart = pd.read_csv('rb_depth_chart.csv')

        
# Merge pff_rushing with rb_depth_chart

pff_rushing = pd.merge(left=pff_rushing, right=rb_depth_chart, how='left', left_on=['Season', 'lower_name'], right_on=['Season', 'Name'], suffixes=[None, '_copy'])
pff_rushing = pff_rushing.drop(columns=['lower_name', 'Team', 'Name'])

pff_rushing
```

Reading and Preparing pff_ol_run data

```python
pff_ol_run_14_15 = pd.read_csv('../fantasy_data/pff_ol_run/pff_ol_run_14-15.csv')
pff_ol_run_15_16 = pd.read_csv('../fantasy_data/pff_ol_run/pff_ol_run_15-16.csv')
pff_ol_run_16_17 = pd.read_csv('../fantasy_data/pff_ol_run/pff_ol_run_16-17.csv')
pff_ol_run_17_18 = pd.read_csv('../fantasy_data/pff_ol_run/pff_ol_run_17-18.csv')
pff_ol_run_18_19 = pd.read_csv('../fantasy_data/pff_ol_run/pff_ol_run_18-19.csv')
pff_ol_run_19_20 = pd.read_csv('../fantasy_data/pff_ol_run/pff_ol_run_19-20.csv')
pff_ol_run_20_21 = pd.read_csv('../fantasy_data/pff_ol_run/pff_ol_run_20-21.csv')
pff_ol_run_21_22 = pd.read_csv('../fantasy_data/pff_ol_run/pff_ol_run_21-22.csv')
pff_ol_run_22_23 = pd.read_csv('../fantasy_data/pff_ol_run/pff_ol_run_22-23.csv')
pff_ol_run_23_24 = pd.read_csv('../fantasy_data/pff_ol_run/pff_ol_run_23-24.csv')
pff_ol_run_24_25 = pd.read_csv('../fantasy_data/pff_ol_run/pff_ol_run_24-25.csv')

pff_ol_run_14_15['Season'] = '2014-15'
pff_ol_run_15_16['Season'] = '2015-16'
pff_ol_run_16_17['Season'] = '2016-17'
pff_ol_run_17_18['Season'] = '2017-18'
pff_ol_run_18_19['Season'] = '2018-19'
pff_ol_run_19_20['Season'] = '2019-20'
pff_ol_run_20_21['Season'] = '2020-21'
pff_ol_run_21_22['Season'] = '2021-22'
pff_ol_run_22_23['Season'] = '2022-23'
pff_ol_run_23_24['Season'] = '2023-24'
pff_ol_run_24_25['Season'] = '2024-25'

pff_ol_run = pd.concat([pff_ol_run_14_15, pff_ol_run_15_16, pff_ol_run_16_17, pff_ol_run_17_18, pff_ol_run_18_19, pff_ol_run_19_20, pff_ol_run_20_21, pff_ol_run_21_22, pff_ol_run_22_23, pff_ol_run_23_24, pff_ol_run_24_25])

# Getting all team names

teams = pff_ol_run['team_name'].unique()

# Getting all seasons

seasons = ['2014-15', '2015-16', '2016-17', '2017-18', '2018-19', '2019-20', '2020-21', '2021-22', '2022-23', '2023-24', '2024-25']

# Calculating total team run block grade from individual player run block grades, weighing in number of run plays for each player
data_rows = []
for season in seasons:
    for team in teams:
        team_df = pff_ol_run[(pff_ol_run['team_name'] == team) & (pff_ol_run['Season'] == season)]
        team_df = team_df[team_df['position'].isin(['C', 'T', 'G'])]

        # Calculating overall grade

        overall_df = team_df[team_df['grades_run_block'].notnull()]
        total_run_snaps = overall_df['snap_counts_run_block'].sum()
        overall_grade = 0
        for index, row in overall_df.iterrows():
            overall_grade += row['grades_run_block'] * (row['snap_counts_run_block'] / total_run_snaps)

        # Calculating overall zone grade by taking in each player's grade and weighing in number of snaps

        zone_df = team_df[team_df['zone_grades_run_block'].notnull()]
        total_zone_snaps = zone_df['zone_snap_counts_run_block'].sum()
        overall_zone_grade = 0
        for index, row in zone_df.iterrows():
            overall_zone_grade += row['zone_grades_run_block'] * (row['zone_snap_counts_run_block'] / total_zone_snaps)

        # Calculating gap grade

        gap_df = team_df[team_df['gap_grades_run_block'].notnull()]
        total_gap_snaps = gap_df['gap_snap_counts_run_block'].sum()
        overall_gap_grade = 0
        for index, row in gap_df.iterrows():
            overall_gap_grade += row['gap_grades_run_block'] * (row['gap_snap_counts_run_block'] / total_gap_snaps)

        row = [team, season, overall_grade, overall_zone_grade, overall_gap_grade]

        # Applying condition so old/new teams (like OAK/LV) don't have a row for every year

        if ((overall_grade != 0) & (overall_zone_grade != 0) & (overall_gap_grade != 0)):
            data_rows.append(row)

        


# Define column names

col_names = ['team_name', 'Season', 'ol_run_block_grade', 'ol_zone_run_block_grade', 'ol_gap_run_block_grade']

# Create pandas dataframe

pff_ol_run = pd.DataFrame(data_rows, columns=col_names)
pff_ol_run
```

Creating df from pfr_fp_final, and adding on the previous 2 seasons of ppg data by joining back to itself

```python
# Start off with pfr_fp_final without including the oldest year in the data

df = pfr_fp_final[pfr_fp_final['Season'] != '2014-15']

# Create column for prev_season

season_key = {'2024-25': '2023-24','2023-24': '2022-23', '2022-23':'2021-22', '2021-22':'2020-21', '2020-21':'2019-20', 
              '2019-20':'2018-19', '2018-19':'2017-18', '2017-18':'2016-17', '2016-17':'2015-16', 
              '2015-16':'2014-15'}

df['prev_season'] = df['Season'].map(season_key)

# Create column for prev_2_season

season_key = {'2024-25': '2022-23','2023-24': '2021-22', '2022-23':'2020-21', '2021-22':'2019-18', '2020-21':'2018-19', 
              '2019-20':'2017-18', '2018-19':'2016-17', '2017-18':'2015-16', '2016-17':'2014-15'}

df['prev_2_season'] = df['Season'].map(season_key)

# Joining df to previous season's pfr_fp_prev to get last season's PPR points

pfr_fp_prev['PPR/G_prev'] = pfr_fp_prev['PPR'] / pfr_fp_prev['G']
pfr_fp_prev = pfr_fp_prev[['PPR/G_prev', 'Season', '-9999']]

df = pd.merge(left=df, right=pfr_fp_prev, how='left', left_on=['-9999', 'prev_season'], right_on=['-9999', 'Season'], suffixes=[None, '_copy'])
df = df.drop(columns=['Season_copy'])

# Joining df to 2 season's ago pfr_fp_prev2 to get PPR points from 2 seasons ago

pfr_fp_prev2['PPR/G'] = pfr_fp_prev2['PPR'] / pfr_fp_prev2['G']
pfr_fp_prev2 = pfr_fp_prev2.rename(columns={"PPR/G":"PPR/G_prev2"})
pfr_fp_prev2 = pfr_fp_prev2[['PPR/G_prev2', 'Season', '-9999']]

df = pd.merge(left=df, right=pfr_fp_prev2, how='left', left_on=['-9999', 'prev_2_season'], right_on=['-9999', 'Season'], suffixes=[None, '_copy'])
df = df.drop(columns=['Season_copy'])
```

Joining Tables Together

```python
# Data Model
# pfr_fp (Currently in df): Fantasy Points from Current Year
# pff_rushing: Rushing Statistics from Previous Year
# pff_receiving: Receiving Statistics from Previous Year
# pff_screens: Screen Receiving Statistics from Previous Year
# pfr_rz_rushing: Red Zone Rushing Statistics from Previous Year
# pfr_tm_off: Current Year team's Offensive Statistics from Previous Year
# pff_ol_pass: Current Year team's OL Statistics from Previous Year


# Joining df (which currently just has pfr data) with current season pff_rushing just to get player_id for future joins. This will only work
# for players that were on one team the whole season.

player_ids = pff_rushing_final[['player', 'player_id', 'Season', 'team_name']]
df = pd.merge(left=df, right=player_ids, how='left', left_on=['Player', 'Tm', 'Season'], right_on=['player', 'team_name', 'Season'], suffixes=[None, '_copy'])
df = df.drop(columns=['player', 'team_name'])

# Join df with current season pff_rushing data without using teams for players that were on multiple teams.
# This will fill in the player_id information for players that were on multiple teams and will do nothing for
# all other players.

df = pd.merge(left=df, right=player_ids, how='left', left_on=['Player', 'Season'], right_on=['player', 'Season'], suffixes=[None, '_copy'])
df['player_id'].fillna(df['player_id_copy'], inplace=True)
df = df.drop(columns=['player_id_copy', 'player', 'team_name'])

# Joining df with with pff_rushing data to get previous year's rushing data

df = pd.merge(left=df, right=pff_rushing_final, how='left', left_on=['player_id', 'prev_season'], right_on=['player_id', 'Season'], suffixes=[None,'_copy'])
df = df.drop(columns=['Season_copy'])
df = df.rename(columns={'team_name':'prev_team'})

# Joining df with with pff_receiving data to get previous year's receiving data

df = pd.merge(left=df, right=pff_receiving_final, how='left', left_on=['player_id', 'prev_season'], right_on=['player_id', 'Season'], suffixes=[None,'_copy'])
df = df.drop(columns=['Season_copy'])

# Joining df with with pff_screens data to get previous year's screen receiving data

df = pd.merge(left=df, right=pff_screens_final, how='left', left_on=['player_id', 'prev_season'], right_on=['player_id', 'Season'], suffixes=[None,'_copy'])
df = df.drop(columns=['Season_copy'])


# Joining df with pfr_rz_rushing data to get previous year's redzone data

df = pd.merge(left=df, right=pfr_rz_rushing_final, how='left', left_on=['-9999', 'prev_season'], right_on=['-9999', 'Season'], suffixes=[None, '_copy'])
df = df.drop(columns=['Season_copy', 'Tm_copy'])

# Joining df with pff_ol_run to get the grade of the current team offensive line's run blocking in the previous season

df = pd.merge(left=df, right=pff_ol_run_final, how='left', left_on=['Tm', 'prev_season'], right_on=['team_name', 'Season'], suffixes=[None, '_copy'])

# Joining df with pfr_tm_off to get statistics about the current team's offense from previous season

df = pd.merge(left=df, right=pfr_tm_off_final, how='left', left_on=['Tm', 'prev_season'], right_on=['Tm', 'Season'], suffixes=[None, '_copy'])
df = df.drop(columns=['Season_copy'])

df.to_csv('rb_df.csv')
df
```
