Scraping Depth Chart Data

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ourlads.com/nfldepthcharts/archive/275/CHI'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')

print(soup.prettify())

# Example: Find all team names
team_names = soup.find_all('div', class_='team-name')
for team in team_names:
    print(team.text.strip())
```

```python
# URL of the page
url = 'https://www.ourlads.com/nfldepthcharts/archive/275/CHI'

# Send a GET request to the URL
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
response = requests.get(url, headers=headers)
```

```python
# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the <td> element with class "row-dc-grey" containing "QB"
    qb_td = soup.find('td', class_='row-dc-grey', string='QB')
    print(qb_td)
    if qb_td != None:
        # Navigate to the next <td> element to get the QB's name
        qb_name_td = qb_td.find_next_sibling('td')
        qb_name_td = qb_name_td.find_next_sibling('td')
        print(qb_name_td)
        if qb_name_td:
            qb_name = qb_name_td.get_text(strip=True)
            print("QB:", qb_name)
        else:
            print("QB name not found")
    else:
        print("QB field not found")
else:
    print("Failed to retrieve the webpage")
```
